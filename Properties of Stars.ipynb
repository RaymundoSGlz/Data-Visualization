# Assessment Part 1: Properties of Stars

```r
library(tidyverse)
library(dslabs)
data(stars)
options(digits = 3)
```

```r
names(stars)
```

```r
## Media de magnitude
mean(stars$magnitude)
```

```r
## Desviación estándar
sd(stars$magnitude)
```

```r
## gráfico de densidad para magnitude
stars |>
    ggplot(aes(magnitude)) +
    geom_density(fill = "gray")
```

```r
## Distribución de temperature
stars |>
    ggplot(aes(temp)) +
    geom_density(fill = "gray")
```

```r
## Scatterplot temperature vs magnitude
stars |>
    ggplot(aes(temp, magnitude)) +
    geom_point()
```

```r
## Aplicando scale_y_reverse
stars |>
    ggplot(aes(temp, magnitude)) +
    geom_point() +
    scale_x_log10() +
    scale_y_reverse()
```

```r
# Filtrar el conjunto de datos para incluir solo estrellas DA, DF y DB
enanas_blancas <- filter(stars, type %in% c("DA", "DF", "DB"))

# Contar el número de filas (que corresponde al número de estrellas enanas blancas)
n_enanas_blancas <- nrow(enanas_blancas)

# Imprimir el número de enanas blancas
cat("Hay", n_enanas_blancas, "estrellas enanas blancas en nuestra muestra.")
```

```r
stars |>
    filter(temp > 5000) |>
    ggplot(aes(temp, magnitude, color = type)) +
    geom_point() +
    geom_text(aes(label = star), nudge_x = 0.075)
```

```r
stars |>
    filter(temp <= median(temp)) |>
    ggplot(aes(temp, magnitude, color = type)) +
    geom_point() +
    geom_text(aes(label = star), nudge_x = 0.075)
```

```r
stars |>
    ggplot(aes(temp, magnitude, color = type)) +
    geom_point() +
    scale_color_manual(values = c("#0072B2", "#E69F00", "#F0E442", "#009E73", "#56B4E9",
                "#D55E00", "#CC79A7", "#999999", "#000000", "#F1C40F"))
```
